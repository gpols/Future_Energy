## Introduction

This research focuses on the relationship between climate variability and energy consumption and production in the United Kingdom over the next 74 years (2024-2098). Each year, rising energy bills place additional financial pressure on households and businesses, making it essential to understand the factors driving these increases. As we face growing concerns about climate change and transition to renewable energy, this study highlights how climate factors-such as air temperature, precipitation, and wind speed-impact electricity demand and renewable energy generation.
The study utilizes data from the [Copernicus Climate Change Service (C3S)](https://cds.climate.copernicus.eu/datasets/sis-energy-derived-projections?tab=overview) Energy operational service, which integrates climate-relevant indicators and energy variables. By analyzing projections, this research aims to clarify how future climate trends may impact energy availability, costs, and our capacity to effectively plan for sustainable energy systems.

The energy sector plays a vital role in modern life, and its operations are increasingly affected by changes in climate patterns [(Schaeffer et al., 2012; IPCC, 2014)](https://www.ipcc.ch/site/assets/uploads/2018/02/ipcc_wg3_ar5_full.pdf). As the UK aims to achieve net-zero emissions by 2050, renewable energy sources such as wind and solar power are expected to become increasingly important (Committee on Climate Change, 2020). However, the variability of these energy sources, influenced by changing weather patterns, presents challenges for maintaining a stable and affordable energy supply [(Bloomfield et al., 2018)](https://essd.copernicus.org/articles/14/2749/2022/essd-14-2749-2022.html).
Studies have shown that temperature changes can directly affect electricity demand [(Zhou & Li, 2017)](https://www.sciencedirect.com/science/article/pii/S0301421520305966), while wind and solar power generation depend heavily on meteorological conditions [(Pfenninger & Staffell, 2016)](https://www.sciencedirect.com/science/article/pii/S1364032119302862). Visualizing these connections through data can help us anticipate potential risks and make informed decisions to safeguard energy security and financial stability.

Recent technological advancements offer promising solutions to energy challenges. An episode of ["Climate Rising"](https://www.hbs.edu/environment/podcast/Pages/podcast-details.aspx?episode=1386170784) by Harvard Business School discusses how artificial intelligence is optimizing energy consumption and reducing costs by analyzing climate and energy data. These technologies enhance grid management, predict energy demands, and improve the integration of renewable energy sources.

The episode also promotes the use of electric vehicles and heat pumps as a means to lower carbon emissions and better manage energy loads. Insights from the discussion reveal that a small percentage of customers account for a significant portion of energy usage, highlighting the need for targeted strategies. These AI-driven approaches enable utilities and consumers to make informed decisions based on their energy consumption timing while emphasizing the importance of adapting to changing energy patterns resulting from climate variability.

The UK is currently grappling with soaring energy prices, with households spending an average of Â£2,500 annually, 91% more than in 2021. This spike is influenced by factors like COVID-19 and the conflict in Ukraine.

The issue stems from the UK's "Marginal Pricing" model, where energy costs are determined by the most expensive source, usually gas, even if cheaper renewables are available. As a result, UK energy prices are much higher than in other countries, with industrial costs being 299% greater than in the USA.

The high prices are exacerbated by government policies, including limited effectiveness of price caps and diminished investment in renewable energy, which maintains reliance on costly energy sources and continues the cycle of high costs.

```python
# Libraries

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualizing temperature data

temp_path = "data_files/air_temperature.csv"

# Skip metadata rows and read the data
temp_df = pd.read_csv(
    temp_path, 
    skiprows=52,  # Skip rows up to the actual data header 
    sep=',',      # Specify the correct delimiter (comma in this case)
    engine='python',  # Use Python engine for flexible parsing
    on_bad_lines='skip'  # Skip any problematic rows
)

# Print the first few rows of the dataframe
print(temp_df.head()) # Up to the first 50 rows are just metadata
print(temp_df.columns) # dtypes: object
print(temp_df.shape) #(593, 39)
```

```python
# First we need to convert the data to a more usable format
# Ensure 'Date' is parsed as a datetime column
temp_df['Date'] = pd.to_datetime(temp_df['Date'], errors='coerce')

# Get the period time range
# Filter rows between 2024 and 2098
temp_pred = temp_df[temp_df['Date'].dt.year >= 2024]

# Return a view instead of copy. Required for python 3.0
# https://pandas.pydata.org/docs/user_guide/copy_on_write.html
pd.options.mode.copy_on_write = True

# Select only 'Date' and 'UK' columns
temp_uk = temp_pred[['Date', 'UK']]

# Convert temperature from Kelvin to Celsius without decimal points
temp_uk['Temperature(Â°C)'] = (temp_uk['UK'] - 273.15).astype(int)

# Drop the original 'UK' column
temp_uk = temp_uk.drop(columns=['UK'])

# Reset the index
temp_uk = temp_uk.reset_index(drop=True)

# Display the resulting data
print(temp_uk)
print(temp_uk.dtypes) # Date: datetime64[ns], Temperature: int64
```

```python
# Load data
energy_path = "data_files/eletricity_demand.csv"

# Skip metadata rows and read the data
energy_df = pd.read_csv(
    energy_path, 
    skiprows=52,  # Skip rows up to the actual data header
    sep=',',      # Specify the correct delimiter (comma in this case)
    engine='python',  # Use Python engine for flexible parsing
    on_bad_lines='skip'  # Skip any problematic rows
)

# Ensure 'Date' is parsed as a datetime column
energy_df['Date'] = pd.to_datetime(energy_df['Date'], errors='coerce')

# Filter rows between 2024 and 2098
# This file has estorical data from 1972 onwards
energy_pred = energy_df[energy_df['Date'].dt.year.between(2024, 2098)]

# Select only 'Date' and 'UK' columns
energy_uk = energy_pred[['Date', 'UK']]

# Change the column name to 'Demand(MWh)'
energy_uk.rename(columns={'UK': 'Demand(MWh)'}, inplace=True)

# Select every 3 months
energy_uk = energy_uk[energy_uk['Date'].dt.month.isin([2, 5, 8, 1])]

# Remove rows of months with no values
energy_uk = energy_uk[energy_uk['Demand(MWh)'] != 0.0]

# Reset the index
energy_uk = energy_uk.reset_index(drop=True)

# Display the resulting data
print(energy_uk)
```

```python
# Load data
wind_path = "data_files/wind_speed.csv"

# Skip metadata rows and read the data
wind_df = pd.read_csv(
    wind_path, 
    skiprows=52,  # Skip rows up to the actual data header
    sep=',',      # Specify the correct delimiter (comma in this case)
    engine='python',  # Use Python engine for flexible parsing
    on_bad_lines='skip'  # Skip any problematic rows
)

# Ensure 'Date' is parsed as a datetime column
wind_df['Date'] = pd.to_datetime(wind_df['Date'], errors='coerce')

# Filter rows between 2024 and 2098
wind_pred = wind_df[wind_df['Date'].dt.year.between(2024, 2098)]

# Select only 'Date' and 'UK' columns
wind_uk = wind_pred[['Date', 'UK']]

# Rename the 'UK' column to 'Wind Speed (m/s)'
wind_uk.rename(columns={'UK': 'Wind Speed (m/s)'}, inplace=True)

# Select every 3 months
wind_uk = wind_uk[wind_uk['Date'].dt.month.isin([2, 5, 8, 11])]

# Reset the index
wind_uk = wind_uk.reset_index(drop=True)

# Display the resulting data
print(wind_uk)
```

```python
# Load data
wind_power_onshore_path = "data_files/wind_power_onshore.csv"

# Skip metadata rows and read the data
wind_power_onshore_df = pd.read_csv(
    wind_power_onshore_path, 
    skiprows=52,  # Skip rows up to the actual data header
    sep=',',      # Specify the correct delimiter (comma in this case)
    engine='python',  # Use Python engine for flexible parsing
    on_bad_lines='skip'  # Skip any problematic rows
)

# Ensure 'Date' is parsed as a datetime column
wind_power_onshore_df['Date'] = pd.to_datetime(wind_power_onshore_df['Date'], errors='coerce')

# Filter rows between 2024 and 2098
wind_power_onshore_pred = wind_power_onshore_df[wind_power_onshore_df['Date'].dt.year.between(2024, 2098)]

# Select only 'Date' and 'UK' columns
wind_power_onshore_uk = wind_power_onshore_pred[['Date', 'UK']]

# Rename the 'UK' column to 'Wind Power Onshore (MW)'
wind_power_onshore_uk.rename(columns={'UK': 'Wind Power Onshore (MW)'}, inplace=True)

# This file only has data of every february
wind_power_onshore_uk = wind_power_onshore_uk[wind_power_onshore_uk['Date'].dt.month.isin([2])]

# Remove rows of months with no values
wind_power_onshore_uk = wind_power_onshore_uk[wind_power_onshore_uk['Wind Power Onshore (MW)'] != 0.0]

# Reset the index
wind_power_onshore_uk = wind_power_onshore_uk.reset_index(drop=True)

# Display the resulting data
print(wind_power_onshore_uk)
```

```python
# Load data
solar_power_path = "data_files/solar_power.csv"

# Skip metadata rows and read the data
solar_power_df = pd.read_csv(
    solar_power_path, 
    skiprows=52,  # Skip rows up to the actual data header
    sep=',',      # Specify the correct delimiter (comma in this case)
    engine='python',  # Use Python engine for flexible parsing
    on_bad_lines='skip'  # Skip any problematic rows
)

# Ensure 'Date' is parsed as a datetime column
solar_power_df['Date'] = pd.to_datetime(solar_power_df['Date'], errors='coerce')

# Filter rows between 2024 and 2098
solar_power_pred = solar_power_df[solar_power_df['Date'].dt.year.between(2024, 2098)]

# Select only 'Date' and 'UK' columns
solar_power_uk = solar_power_pred[['Date', 'UK']]

# Rename the 'UK' column to 'Solar Power (MW)'
solar_power_uk.rename(columns={'UK': 'Solar Power (MW)'}, inplace=True)

# Select every 3 months
solar_power_uk = solar_power_uk[solar_power_uk['Date'].dt.month.isin([2, 5, 8, 11])]

# Remove rows of months with no values
solar_power_uk = solar_power_uk[solar_power_uk['Solar Power (MW)'] != 0.0]

# Reset the index
solar_power_uk = solar_power_uk.reset_index(drop=True)

# Display the resulting data
print(solar_power_uk)
```

The energy demand data is presented in MWh, which is not directly comparable to the solar power data in MW.

To make a meaningful comparison, we need to convert the energy demand data to MW.

```python
# Convert energy demand from MWh to MW
# Since the measure is every 3 months 
# Power (MW) = Energy (MWh) / Time (hours)
# 3 months is 2160 hours

energy_uk_converted = energy_uk.copy()
energy_uk_converted['Demand(MW)'] = energy_uk_converted['Demand(MWh)'] / 2160

# Merge data 
merged_data = solar_power_uk.merge(energy_uk_converted[['Date', 'Demand(MW)']], on='Date')

fig, ax1 = plt.subplots(figsize=(14, 7))

# Line plot for solar power
sns.lineplot(
    data=merged_data, 
    x='Date', 
    y='Solar Power (MW)', 
    ax=ax1, 
    color='salmon', 
    linewidth=2, 
    marker='o'
)
ax1.set_ylabel('Solar Power (MW)', fontsize=12)
ax1.tick_params(axis='y')

# Line plot for energy demand on the second axis
ax2 = ax1.twinx()
sns.lineplot(
    data=merged_data, 
    x='Date', 
    y='Demand(MW)', 
    ax=ax2, 
    color='skyblue', 
    linewidth=2, 
    marker='o'
)
ax2.set_ylabel('Energy Demand (MW)', fontsize=12)
ax2.tick_params(axis='y')

# Labels and Title
ax1.set_xlabel('Date', fontsize=12)
ax1.set_title('Solar Power x Energy Demand (MW)', fontsize=14)

ax1.legend(['Solar Power'], loc='upper left')
ax2.legend(['Energy Demand'], loc='upper right')
ax1.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python

# Calculate the price in Â£ MWh
energy_uk['Price(Â£)'] = energy_uk['Demand(MWh)'] *  248.60 

# How much it will increase?
print(energy_uk)

# Find peak demand (taking top 10% highest demand)
peak_demand = energy_uk.nlargest(int(len(energy_uk) * 0.1), 'Demand(MWh)')

# Set seaborn theme for clarity
sns.set_theme(style="darkgrid")

# Plotting the data
fig, ax1 = plt.subplots(figsize=(14, 7))

# Time series plot for energy demand with error bands
sns.lineplot(
    data=energy_uk, 
    x='Date', 
    y='Demand(MWh)', 
    ax=ax1, 
    color='skyblue', 
    linewidth=7,
    alpha=0.7,
    marker='o',
    label="Energy Demand"
)
ax1.set_ylabel('Energy Demand (MWh)', fontsize=12)
ax1.tick_params(axis='y')

# Time series plot for price on the second axis with error bands
ax2 = ax1.twinx()
sns.lineplot(
    data=energy_uk, 
    x='Date', 
    y='Price(Â£)', 
    ax=ax2, 
    color='salmon', 
    linewidth=2,
    marker='o',
    label="Price (Â£)"
)
ax2.set_ylabel('Price (Â£)', fontsize=12)
ax2.tick_params(axis='y')

# Highlight peak demand points with red scatter plot
sns.scatterplot(
    data=peak_demand, 
    x='Date', 
    y='Demand(MWh)', 
    color='red', 
    s=100, 
    label='Peak Demand', 
    ax=ax1
)

# Labels and Title
ax1.set_xlabel('Date', fontsize=12)
ax1.set_title('Energy Demand (MWh) x Price (Â£)', fontsize=14)

# Legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Grid and layout
ax1.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()

# Knowing the average price of energy and the price during peak demand periods can help us understand the cost implications of high demand.
# Calculate the average price of energy
average_price = energy_uk['Price(Â£)'].mean()

# Calculate the price during peak demand periods
peak_price = peak_demand['Price(Â£)'].mean()

# Calculate how much higher the peak price is compared to the average price
price_difference = peak_price - average_price

# Calculate the percentage increase in price during peak demand
percentage_increase = (price_difference / average_price) * 100
 
print(f"Price increase during peak demand: {percentage_increase:.2f}%")
```

The next cell will analyze the cost implications of wind power generation.
Identify peak wind power periods and compare the cost per hour during these peak periods to the average cost per hour. 

How much will the price increase during peak demand?


```python
# The original data is for 3 months (2160 hours)
# Copy the wind power data
wind_power = wind_power_onshore_uk.copy()

# Calculate the total energy produced in MWh for each 3-month period
# Energy produced = Power (MW) * Time (hours) = MW * 2160 hours
wind_power['Energy (MWh)'] = wind_power['Wind Power Onshore (MW)'] / 2160

# Now, calculate the cost for the 3-month period
wind_power['Cost(Â£)'] = wind_power['Energy (MWh)'] * 63

# Calculate the cost per hour by dividing by 2160 (the number of hours in 3 months)
wind_power['Cost per Hour (Â£)'] = wind_power['Cost(Â£)'] / 2160

# Find peak wind power (taking top 10% highest wind power)
peak_wind_power = wind_power.nlargest(int(len(wind_power) * 0.1), 'Wind Power Onshore (MW)')

# Set seaborn theme for clarity
sns.set_theme(style="darkgrid")

# Plotting the data
fig, ax1 = plt.subplots(figsize=(14, 7))

# Time series plot for wind power with error bands
sns.lineplot(
    data=wind_power, 
    x='Date', 
    y='Wind Power Onshore (MW)', 
    ax=ax1, 
    color='skyblue', 
    linewidth=7,
    alpha=0.7,
    marker='o',
    label="Wind Power"
)

ax1.set_ylabel('Wind Power Onshore (MW)', fontsize=12)
ax1.tick_params(axis='y')

# Time series plot for cost per hour with error bands
ax2 = ax1.twinx()
sns.lineplot(
    data=wind_power, 
    x='Date', 
    y='Cost per Hour (Â£)', 
    ax=ax2, 
    color='salmon', 
    linewidth=2,
    marker='o',
    label="Cost per Hour (Â£)"
)

ax2.set_ylabel('Cost per Hour (Â£)', fontsize=12)
ax2.tick_params(axis='y')

# Highlight peak wind power points with red scatter plot
sns.scatterplot(
    data=peak_wind_power, 
    x='Date', 
    y='Wind Power Onshore (MW)', 
    color='red', 
    s=100, 
    label='Peak Wind Power', 
    ax=ax1
)

# Labels and Title
ax1.set_xlabel('Date', fontsize=12)
ax1.set_title('Wind Power (MW) x Cost per Hour (Â£)', fontsize=14)

# Legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Grid and layout
ax1.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()

# Calculate the average cost per hour of wind power
average_cost_per_hour = wind_power['Cost per Hour (Â£)'].mean()

# Calculate the cost per hour during peak wind power periods
peak_cost_per_hour = peak_wind_power['Cost per Hour (Â£)'].mean()

# Calculate how much higher the peak cost is compared to the average cost
cost_difference_per_hour = peak_cost_per_hour - average_cost_per_hour

# Calculate the percentage increase in cost during peak wind power
percentage_increase_per_hour = (cost_difference_per_hour / average_cost_per_hour) * 100

print(f"Price increase during peak demand: {percentage_increase_per_hour:.2f}%")
```

Solar energy is generally more cost-effective than wind energy, both in terms of upfront investment and maintenance. Solar panel systems are typically cheaper, with average costs around Â£7,026 for a three-bedroom house. 

In contrast, wind turbines, especially standalone ones, can range from Â£7,000 to Â£70,000, making them a significantly larger financial commitment. Additionally, solar panels tend to have lower maintenance costs compared to wind turbines, which require more upkeep due to moving parts and the complexities of installation. 

As a result, solar energy offers a more affordable and accessible option for many households compared to wind energy.

But how much energy each of them will produce?

```python
# Merge the solar power and wind power data
merged_power = solar_power_uk.merge(wind_power[['Date', 'Wind Power Onshore (MW)']], on='Date')

# Normalize both columns
merged_power['Solar Power'] = merged_power['Solar Power (MW)'] / merged_power['Solar Power (MW)'].max()
merged_power['Wind Power'] = merged_power['Wind Power Onshore (MW)'] / merged_power['Wind Power Onshore (MW)'].max()

# Reshape to long-form
melt_power = merged_power.melt(
    id_vars="Date", 
    value_vars=["Solar Power", "Wind Power"], 
    var_name="Power Source", 
    value_name="Power"
)

# Plot
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=melt_power,
    x="Date",
    y="Power",
    hue="Power Source",
    style="Power Source",
    markers=True,
    linewidth=2
)

# Add labels and title
plt.title("Solar x Wind Power Generation", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Power", fontsize=12)

# Adjust legend and grid
plt.legend(title="Power Source", loc="upper left")
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.tight_layout()
plt.show()
```

Additionally, improved forecasting can help policymakers create incentives to promote energy efficiency during high-demand periods, further alleviating financial strain on households.
